#QUANTIDADE DE MUNICÍPIOS POR ESTADO

```python
import matplotlib.pyplot as plt

municipios_por_estado = dados['estado'].value_counts().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
municipios_por_estado.plot(kind='bar')
plt.xlabel('Estado')
plt.ylabel('Número de municípios')
plt.title('Quantidade de municípios por estado')
plt.xticks(rotation=90)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

#Mapa de dispersão por latitude/longitude (pontos)

```python
plt.figure(figsize=(8, 8))
plt.scatter(dados['longitude'], dados['latitude'], s=5, alpha=0.5)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Localização dos municípios brasileiros')
plt.grid(True, alpha=0.3)
plt.show()
```

#Municipios por região

```python
municipios_por_uf = dados['uf'].value_counts()

plt.figure(figsize=(10, 5))
municipios_por_uf.plot(kind='bar', color='teal')
plt.xlabel('UF')
plt.ylabel('Número de municípios')
plt.title('Quantidade de municípios por UF')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

#Gráfico de barras: capitais x não capitais

```python
capitais = dados['capital'].value_counts()
capitais.index = ['Não capital', 'Capital']  # ajuste conforme valores

plt.figure(figsize=(5, 4))
capitais.plot(kind='bar', color=['gray', 'orange'])
plt.ylabel('Quantidade de municípios')
plt.title('Capitais x não capitais')
plt.grid(axis='y', alpha=0.3)
plt.show()
```
